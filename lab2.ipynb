## Лабораторная работа 2

Одноступенчатая ракета, необходимая для вывода на орбиту Земли искусственного спутника, и многоступенчатая ракета.

$V$ - скорость истечения сгорающего топлива 3 км/с;

$m_0$ - масса ракеты на старте; 

$m_p$ - "полезная" нагрузка;

$\lambda$ - коэффициент структурной массы;

Необходимо вычислить $m_s, m_t$ - массу топлива и выявить зависимость скорости ракеты от количества ступеней, рассчитать, сколько ступеней эффективнее всего использовать.

# Одноступенчатая ракета

Рассмотрим идеальный обьект

$\alpha = \frac{V^2}{R_з+h}$

$\lambda = \frac{m_s}{m_s+m_t} = 0.1$ - коэффицент структурной массы

Гравитации и сопративления воздуха - нет, $u$ - скорость сгорающего топлива $3 км\с$ - константа, скорость вращения земли так же не учитывается.

Имеется: одноступенчатая ракета. Масса ракетной конструкции, которая не может быть включена в состав ступени, включается в $m_p$ , так вместе с полезным грузом выводится на орбиту.

Расчет:

Для начала учтем, что на ракету, двигающуюся по орбите с радиусом $R_з + h$ действует сила притяжения Земли и сила, обусловленная вторым законом Ньютона.

$m_c\frac{V^2}{R_з+h} = G\frac{m_c M_з}{(R_з+h)^2}$

из этого следует, что для первой космической скорости нужно

$V = \sqrt{\frac{GM_з}{R_з+h}} \approx \sqrt{\frac{6.67*10^{-11} * 5.97*10^24}{6371*10^3}} \approx 7.9 км/с$

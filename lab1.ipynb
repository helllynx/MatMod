## Модель 1

**Стрельба в горизонтальном направлении с измерением дальности полета пули**

На высоте $h$ параллельно горизонту закреплено ружье. После выстрела уля приземляется на расстоянии $L$ от стенда с ружьем. Считаем, что нет трения пули с воздухом, земля плоская, никакие силы кроме тяжести не действуют на пулю.

![Pic1](https://preview.ibb.co/hd0azQ/exp1.png "Pic 1")


$V = \sqrt{\frac{L^2g}{2h}}$

$L_0 = \sqrt{\frac{2v_0^2h}{g}}$

Проведем эксперимент, попробуем оценить среднюю скорость пули.

Для достоверности расчетов, учтем случайные факторы, будем считать L нормально распределенной случайной величиной с мат. ожиданием $360 м$ и $\sigma = L_0*0.05 м$

## Модель 2

**(Закон сохранения энергии)**

Cтенд для измерения скорости пули, включающий ружье, закрепленное на горизонтальной поверхности. Стрельба производится в ящик с песком.

1. Ящик подвешен на нерастяжимом стержне;
2. Масса ящика M во много раз больше массы пули m;
3. Ящик и пуля рассматриваются как система материальных точек;
4. Ящик расположен близко к дулу и сопротивление воздуха пренебрежимо мало;
5. Стрельба происходит без отдачи;
6. Трение точки опоры пренебрежимо мало;
7. Вся кинетическая энергия пули полностью переходит в потенциальную энергию системы «пуля – ящик», пренебрегая переходом части кинетической энергии пули в тепловую.

![Pic1](https://preview.ibb.co/g5bnDk/exp2.png "Pic 2")

Искомую скорость и угол мы найдем по формуле:

$V = \sqrt{2gl(1-cos(\alpha))\frac{M+m}{m}}$

$\alpha_0 = arccos(\frac{-V_0^2m}{2gl(M+m)}+1)$

Предположим, что если не учитываемые нами потери энергии будут невелики, то скорость, полученная при вычислениях по этой формуле, будет вполне точной.
Однако, процессы, происходящие при «слипании» пули и маятника, уже не являются чисто механическими. Поэтому примененный для вычисления величины V закон сохранения механической энергии несправедлив: сохраняется полная, а не механическая энергия системы. Он дает лишь нижнюю границу для оценки скорости пули.

Если проведем расчеты, то получим странные результаты(Масса пули $m=0.01kg$, груза $M=1000kg$):

Получаем неадекватные значения:

$\alpha_0 = 0.579500421471148$

Что в переводе на градусы $33.202928376431536606$.
Как то очень странно. Не правда ли?) (Хотел бы посмотреть, как пуля сдвинет груз с таким весом на 33 градуса)

Значит, модель некорректна.

## Модель 3

**(Закон сохранения импульса)**

Cтенд для измерения скорости пули, включающий ружье, закрепленное на горизонтальной поверхности. Стрельба производится в ящик с песком.

![Pic1](https://preview.ibb.co/g5bnDk/exp2.png "Pic 2")

Искомую скорость и угол мы найдем по формуле:

$V = \frac{M+m}{m}\sqrt{2gl(1-cos(\alpha))}$

$\alpha_0=arccos(1-\frac{1}{2gl}(\frac{v_0 m}{(M+m)})^2)$

Если проведем расчеты, то получим (Масса пули $m=0.01kg$, груза $M=1000kg$):

$\alpha_0 = 0.0018069979816725296$

Что в переводе на градусы $0.10353335793875084736$.
